```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

def plot_dataframe(df,data_type="numerical"):
    """
    Plots the given DataFrame columns in a grid layout. Chooses the plot type based on data type.

    Parameters:
    - df (DataFrame): DataFrame containing the data to plot.
    - columns (list): List of column names to plot.
    - data_type (str): Type of data to plot ("numerical" or "categorical").
    """
    # Set style for a clean look
    sns.set(style="whitegrid")

    # Determine the number of rows and columns for subplots based on the number of columns to plot)
    if data_type=='numerical':
        grid_size=3
    elif data_type=='categorical':
        grid_size=2
    fig, ax = plt.subplots(grid_size, grid_size, figsize=(20,15))
    ax = ax.flatten()  # Flatten axes for easy iteration

    # Loop through each column and create the appropriate plot type
    for i, col in enumerate(df.columns):
        if data_type == "numerical":
            # KDE plot for numerical columns
            sns.kdeplot(data=df[col].dropna(), ax=ax[i], fill=True, color='skyblue', linewidth=2, alpha=0.6)
        elif data_type == "categorical":
            # Count plot for categorical columns
            sns.countplot(data=df, x=col, ax=ax[i], palette="plasma")
        
        # Set titles and labels
        ax[i].set_title(col, fontsize=16, fontweight='bold')
        ax[i].set_xlabel('Value' if data_type == "numerical" else col, fontsize=14)
        ax[i].set_ylabel('Density' if data_type == "numerical" else 'Frequency', fontsize=14)
        
        # Customize the grid and ticks
        ax[i].grid(True, linestyle='--', alpha=0.5)
        ax[i].tick_params(axis='both', which='major', labelsize=12)
        ax[i].tick_params(axis='x', rotation=45)
    # Hide any unused axes if there are extra subplots
    for j in range(i + 1, len(ax)):
        fig.delaxes(ax[j])

    # Adjust layout for better spacing
    plt.tight_layout(pad=3.0)
    
    # Save the figure
    plt.savefig(f'data/high_quality_plot_{data_type}.png', dpi=300, bbox_inches='tight')
    
    # Show the plot
    plt.show()


```

```python
def age_categories(age: int) -> str:
    """
    Categorizes a given age into one of three predefined age groups.

    Args:
        age (int): The age of an individual. Must be a positive integer.

    Returns:
        str: A string representing the age category:
            - 'UnderAge' if the age is less than 23.
            - 'HealthyAge' if the age is between 23 (inclusive) and 38 (exclusive).
            - 'OldAge' if the age is 38 or above.

    Example:
        >>> age_categories(25)
        'HealthyAge'

        >>> age_categories(40)
        'OldAge'

        >>> age_categories(20)
        'UnderAge'
    """
    if age < 23:
        return 'UnderAge'
    elif age >= 23 and age < 38:
        return 'HealthyAge'
    else:
        return 'OldAge'
```

```python
new_df['Age_Category']=new_df['Maternal_Age'].apply(age_categories)
new_df.groupby('Age_Category',as_index=False)['Delivery_Mode'].value_counts()
```

```python
import pandas as pd

def healthy_weight(row: float) -> str:
    """
    Categorizes birth weight as 'Healthy' or 'Unhealthy', or identifies missing data.

    Args:
        row (float): The birth weight of the baby. Expected to be a numerical value (in grams).

    Returns:
        str: A string representing the category of the birth weight:
            - 'Healthy' if the weight is between 2550 and 4000 grams (exclusive).
            - 'Unhealthy' if the weight is outside the healthy range.
            - 'Missing Data' if the value is NaN or missing.

    Example:
        >>> healthy_weight(3500)
        'Healthy'

        >>> healthy_weight(4500)
        'Unhealthy'

        >>> healthy_weight(None)
        'Missing Data'
    """
    if pd.isna(row):
        return 'Missing Data'
    elif 2550 < row < 4000:
        return 'Healthy'
    else:
        return 'Unhealthy'


def healthy_height(row: float) -> str:
    """
    Categorizes birth height as 'Healthy' or 'Unhealthy', or identifies missing data.

    Args:
        row (float): The birth height (length) of the baby. Expected to be a numerical value (in centimeters).

    Returns:
        str: A string representing the category of the birth height:
            - 'Healthy' if the height is between 39 and 61 centimeters (exclusive).
            - 'Unhealthy' if the height is outside the healthy range.
            - 'Missing Data' if the value is NaN or missing.

    Example:
        >>> healthy_height(50)
        'Healthy'

        >>> healthy_height(65)
        'Unhealthy'

        >>> healthy_height(None)
        'Missing Data'
    """
    if pd.isna(row):
        return 'Missing Data'
    elif 39 < row < 61:
        return 'Healthy'
    else:
        return 'Unhealthy'


# Example of applying the functions to a DataFrame
df['Weight'] = df['Birth_Weight'].apply(healthy_weight)
df['Height'] = df['Birth_Length'].apply(healthy_height)
```

```python
import statsmodels.api as sm
import pandas as pd

def analyze_association_with_age(df: pd.DataFrame, cols: list) -> None:
    """
    Analyzes the association between maternal age and the likelihood of certain health conditions, including NICU stay, birth weight, and height.
    
    This function performs the following steps:
    1. Cleans the data by removing rows with missing values for the specified columns and maternal age.
    2. Converts the categorical columns into binary values for logistic regression:
        - 'Yes' or 'Unhealthy' values are mapped to 1.
        - 'No' or 'Healthy' values are mapped to 0.
    3. Fits a logistic regression model using maternal age as the predictor and the binary outcome variable.
    4. Prints whether there is a significant association between maternal age and the likelihood of each condition based on p-values.

    Args:
        df (pd.DataFrame): The DataFrame containing the data to be analyzed.
        cols (list): A list of column names to analyze. These should include 'NICU_Stay', 'Weight', and 'Height'.

    Returns:
        None: The function prints the results of the analysis for each column in `cols`.
    
    Example:
        >>> analyze_association_with_age(df, ['NICU_Stay', 'Weight', 'Height'])
        There is significant association between age of pregnant women and likelihood of NICU_Stay
        There is no significant association between age of pregnant women and likelihood of Weight
        There is significant association between age of pregnant women and likelihood of Height
    """
    
    # Ensure NICU_Stay is treated as a categorical variable
    df['NICU_Stay'] = df['NICU_Stay'].astype('object').fillna('Missing Data')
    
    # Iterate over specified columns
    for col in cols:
        # Data cleaning: filter rows with missing data
        temp_df = df[(df[col] != 'Missing Data') & (df['Maternal_Age'].notnull())]
        
        # Binary conversion based on the column
        if col == 'NICU_Stay':
            temp_df['Binary'] = temp_df[col].apply(lambda x: 1 if x == 'Yes' else 0)
        else:
            temp_df['Binary'] = temp_df[col].apply(lambda x: 1 if x == 'Unhealthy' else 0)
        
        # Logistic regression model
        X = temp_df['Maternal_Age']
        y = temp_df['Binary']
        x = sm.add_constant(X)  # Adds a constant (intercept) to the model
        
        logit_model = sm.Logit(y, x)  # Logistic regression model
        result = logit_model.fit()  # Fit the model
        
        # Extract the p-value for Maternal_Age
        p_value = result.pvalues['Maternal_Age']
        alpha = 0.3  # Significance level
        
        # Print whether there is a significant association based on the p-value
        if p_value < alpha:
            print(f"There is significant association between age of pregnant women and likelihood of {col}")
        else:
            print(f"There is no significant association between age of pregnant women and likelihood of {col}")
        
        print()  # Print a blank line for better readability between results
analyze_association_with_age(df,cols)
```

#### Significance Level Chosen: 30%

In statistical analysis, the **significance level** represents the probability threshold at which we accept that our results may contain some degree of error. Typically, a significance level of 5% or 1% is used, meaning we are willing to tolerate a 5% or 1% chance of error, which translates into a 95% or 99% confidence in the validity of our findings.

However, in certain fieldssuch as healthcare or medical researcha 95% confidence level may not always be necessary. Consider this scenario: if we are investigating whether a disease is causing fatalities, would we wait until 9 out of 10 cases result in death to conclude that the disease is lethal? Most likely not. In practice, we would declare the disease to be a significant threat even if it results in 6 out of 10 fatalities.

For this reason, we have opted for a **significance level of 30%**, which allows for a more flexible and timely analysis while acknowledging that some uncertainty in the data is acceptable in certain contexts.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_line_graphs(df: pd.DataFrame, **data: dict) -> None:
    """
    Plots two line graphs for visualizing relationships between maternal age and various factors such as EPDS, anxiety scores, and gestational age.

    This function generates two subplots:
    1. A line plot comparing EPDS scores and anxiety levels with maternal age.
    2. A line plot comparing gestational age at birth with maternal age.

    The function takes in a DataFrame and a dictionary containing the column names for plotting. It uses Seaborn and Matplotlib to generate the plots, enhancing their appearance for better clarity.

    Args:
        df (pd.DataFrame): The DataFrame containing the data to be plotted.
        **data (dict): A dictionary containing the following keys:
            - 'x': The column name for the x-axis (typically 'Maternal_Age').
            - 'y1': The column name for the first y-axis (typically 'EPDS' or any other metric).
            - 'y2': The column name for the second y-axis (typically 'Anxiety' or another related metric).
            - 'y3': The column name for the third y-axis (typically 'Gestational_Age' or other health-related measures).

    Returns:
        None: The function generates and displays the plots.

    Example:
        >>> plot_line_graphs(df, x='Maternal_Age', y1='EPDS', y2='Anxiety', y3='Gestational_Age')
    """
    # Set the Seaborn style and color palette for a cleaner look
    sns.set(style="whitegrid")
    
    # Initialize subplots
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # Widen figure size for better readability
    axs = axs.flatten()
    
    # Plot EPDS and Anxiety vs. Maternal Age
    sns.lineplot(x=data['x'], y=data['y1'], data=df, ax=axs[0], color='red', label=data['y1'], linewidth=2, linestyle='-')
    sns.lineplot(x=data['x'], y=data['y2'], data=df, ax=axs[0], color='blue', label=data['y2'], linewidth=2, linestyle='--')
    
    # Set titles and labels for the first plot
    axs[0].set_title(f"{data['x']} vs. {data['y1']+data['y2']}", fontsize=12, color='green', fontweight='bold')
    axs[0].set_xlabel(data['x'], fontsize=13)
    axs[0].set_ylabel('Scores', fontsize=13)
    
    # Add a legend with a better placement
    axs[0].legend(title=f"{data['y1']} and {data['y2']}", title_fontsize='13', loc='best', fontsize=12, frameon=True)
    
    # Plot Gestational Age at Birth vs. Maternal Age
    sns.lineplot(x=data['x'], y=data['y3'], data=df, ax=axs[1], color='green', label=data['y3'], linewidth=2)
    
    # Set titles and labels for the second plot
    axs[1].set_title(f"{data['x']} vs. {data['y3']}", fontsize=12, color='green', fontweight='bold')
    axs[1].set_xlabel(data['x'], fontsize=13)
    axs[1].set_ylabel(data['y3'], fontsize=13)
    
    # Customize tick parameters for both axes for better readability
    for ax in axs:
        ax.tick_params(axis='both', which='major', labelsize=11)
        ax.grid(True, linestyle='--', alpha=0.6)
    
    # Tight layout and show the plot
    plt.tight_layout(pad=3.0)
    plt.show()
```

```python
import plotly.express as px
import pandas as pd

def bivariate_analysis(df: pd.DataFrame,
                       x: str,
                       feature: str = 'Household_Income',
                       title: str = None,
                       normal_value: int = None) -> None:
    """
    Performs a bivariate analysis by creating a histogram that visualizes the distribution 
    of a given feature (e.g., a variable) categorized by another feature (e.g., Household Income).

    The function also allows for the optional addition of a vertical line representing a 
    "normal value" if provided.

    Args:
        df (pd.DataFrame): The DataFrame containing the data to be plotted.
        x (str): The column name in the DataFrame representing the variable to be analyzed.
        feature (str, optional): The categorical feature by which to group the data (default is 'Household_Income').
        title (str, optional): The title for the x-axis, which can be customized (default is None).
        normal_value (int, optional): A specific value (e.g., normal or threshold value) to mark with a dotted line on the plot (default is None).

    Returns:
        None: The function generates and displays an interactive histogram plot using Plotly.

    Example:
        >>> bivariate_analysis(df, x='Maternal_Age', feature='Household_Income', title='Age Distribution', normal_value=30)
    """
    
    categories = df[feature].unique()

    # Create the figure
    fig = px.histogram(df, x=x, color=feature,
                       title=f'Distribution of {x} by {feature}',
                       hover_data=df.columns)

    # Add a dotted vertical line if normal_value is provided
    if normal_value is not None:
        fig.add_vline(x=normal_value, line=dict(dash='dot', width=2, color='red'), 
                      annotation_text=f"Normal Value: {normal_value}", annotation_position="top right")

    # Update the layout for better visuals
    fig.update_layout(
        xaxis_title=title,
        yaxis_title="Count",
        legend_title=feature,
        barmode='overlay'  # Overlay histograms for better comparison
    )

    # Show the plot
    fig.show()
```

```python
import pandas as pd

def feature_impact(df: pd.DataFrame, y: str, cols: list[str]) -> pd.DataFrame:
    """
    Analyzes the impact of multiple features on a target variable by calculating the mean of 
    each feature grouped by the target variable, and returns a DataFrame with the results.

    Args:
        df (pd.DataFrame): The DataFrame containing the data to be analyzed.
        y (str): The column name of the target variable (the feature used for grouping).
        cols (list[str]): A list of column names whose means will be calculated and compared across 
                          the different values of the target variable (y).

    Returns:
        pd.DataFrame: A DataFrame containing the means of the specified columns, grouped by the 
                      target variable `y`. Each column will have the suffix '_mean' added to 
                      indicate that it represents the mean.

    Example:
        >>> cols = ['Threaten_Life', 'Threaten_Baby_Danger', 'Threaten_Baby_Harm']
        >>> result = feature_impact(df=df, y='Household_Income', cols=cols)
        >>> result
    """
    # Create an empty DataFrame to store the results
    result_df = pd.DataFrame()
    
    # For each column in 'cols', calculate the mean by y and add it to the result dataframe
    for col in cols:
        # Filter the dataframe to remove rows where col is null
        temp_df = df.query(f'{col}.notnull()')
        
        # Calculate the mean for the specified column grouped by y
        temp_result = temp_df.groupby(y)[col].mean().reset_index()
        temp_result = temp_result.rename(columns={col: f'{col}_mean'})  # Rename the column to show it's the mean
        
        # Merge the result into the result_df DataFrame
        result_df = pd.merge(result_df, temp_result, on=y, how='outer') if not result_df.empty else temp_result

    # Sorting the result dataframe by y for better presentation
    result_df = result_df.sort_values(y)

    return result_df
cols = ['Threaten_Life', 'Threaten_Baby_Danger', 'Threaten_Baby_Harm']
result = feature_impact(df=df, y='Household_Income', cols=cols)
result
```

```python
import pandas as pd

def analyze_categorical_vs_numeric(df: pd.DataFrame, 
                                    cat_feature: str, 
                                    num_feature: str, 
                                    threshold: float, 
                                    normal_label: str = "Below Threshold", 
                                    above_label: str = "Above Threshold") -> pd.DataFrame:
    """
    Analyzes the relationship between a categorical feature and a numeric feature by calculating 
    the percentage distribution of values above and below a specified threshold.
    
    Parameters:
    df (pd.DataFrame): The input dataframe containing the data.
    cat_feature (str): The name of the categorical feature in the dataframe.
    num_feature (str): The name of the numeric feature in the dataframe.
    threshold (float): The threshold value to classify the numeric feature.
    normal_label (str): The label for values below the threshold (default is "Below Threshold").
    above_label (str): The label for values above the threshold (default is "Above Threshold").
    
    Returns:
    pd.DataFrame: A dataframe containing the percentage distribution of the categorized numeric feature 
                  across the different levels of the categorical feature.
    
    Example:
        >>> result = analyze_categorical_vs_numeric(df=df, cat_feature='Category', num_feature='Age', threshold=30)
        >>> print(result)
    """
    
    # Create a new column categorizing the numeric feature based on the threshold
    df['Category_Status'] = df[num_feature].apply(lambda x: above_label if x > threshold else normal_label)

    # Group by the categorical feature and the new status to count occurrences
    analysis = (
        df.groupby([cat_feature, 'Category_Status'])
        .size()
        .reset_index(name='Count')  # Count occurrences
        .pivot(index=cat_feature, columns='Category_Status', values='Count')  # Reshape for percentages
        .fillna(0)  # Fill NaN with 0 for missing groups
    )

    # Calculate total counts and percentages for each category
    analysis['Total'] = analysis.sum(axis=1)
    analysis[f'% {normal_label}'] = (analysis[normal_label] / analysis['Total']) * 100
    analysis[f'% {above_label}'] = (analysis[above_label] / analysis['Total']) * 100

    # Final dataframe with meaningful representation
    result_df = analysis[[f'% {normal_label}', f'% {above_label}']].reset_index()
    result_df.columns = [cat_feature, f'% {normal_label}', f'% {above_label}']
    result_df.sort_values(by=f'% {above_label}', ascending=False, inplace=True)
    
    return result_df
```

```python
def calculate_conditional_probability(df, condition_col, condition_thresh, target_col, target_thresh, condition_op="<", target_op=">"):
    """
    Calculate the conditional probability of `target_col` meeting a threshold given 
    `condition_col` meets a condition.
    
    Args:
        df (pd.DataFrame): The DataFrame containing the data.
        condition_col (str): Column for the condition.
        condition_thresh (numeric): Threshold for the condition column.
        target_col (str): Column for the target variable.
        target_thresh (numeric): Threshold for the target column.
        condition_op (str): Operator for the condition ('<', '<=', '>', '>=').
        target_op (str): Operator for the target ('<', '<=', '>', '>=').

    Returns:
        float: The conditional probability (percentage).
    """
    # Define comparison operators
    ops = {
        "<": lambda x, y: x < y,
        "<=": lambda x, y: x <= y,
        ">": lambda x, y: x > y,
        ">=": lambda x, y: x >= y
    }
    
    # Validate operators
    if condition_op not in ops or target_op not in ops:
        raise ValueError("Invalid operator. Use one of: '<', '<=', '>', '>='.")
    
    # Apply conditions
    condition_met = ops[condition_op](df[condition_col], condition_thresh)
    target_met = ops[target_op](df[target_col], target_thresh)
    
    # Calculate probabilities
    total_condition = condition_met.sum()
    joint_condition = (condition_met & target_met).sum()
    
    if total_condition == 0:
        return 0  # Avoid division by zero
    
    probability = (joint_condition / total_condition) * 100
    return probability

# Apply the function for the two cases
low_epds_high_anxiety = calculate_conditional_probability(
    df=df,
    condition_col='EPDS',
    condition_thresh=12,
    target_col='Anxiety',
    target_thresh=20,
    condition_op="<",
    target_op=">"
)

high_epds_high_anxiety = calculate_conditional_probability(
    df=df,
    condition_col='EPDS',
    condition_thresh=12,
    target_col='Anxiety',
    target_thresh=20,
    condition_op=">",
    target_op=">"
)

# Print results
print(f"Probability of women with low EPDS (<12) having high Anxiety (>20): {low_epds_high_anxiety:.2f}%")
print(f"Probability of women with high EPDS (>12) having high Anxiety (>20): {high_epds_high_anxiety:.2f}%")
```

    *   This finding underscores the profound impact of **socioeconomic status** on perceived vulnerability and health security during times of crisis.
        
4.  **Interplay Between Education and Income**
    
    *   Women with a **masters degree but low income** and those with **high income but lower education levels** exhibit similar levels of concern regarding the pandemics effects on maternal and infant health.
        
    *   This highlights that both **education** and **income** are equally important in shaping perceptions of risk and worry, suggesting that neither factor can be viewed in isolation when addressing maternal health concerns.
